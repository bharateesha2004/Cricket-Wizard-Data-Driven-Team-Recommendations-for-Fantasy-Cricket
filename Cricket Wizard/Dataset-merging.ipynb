```python
import pandas as pd

# Reading CSV files into DataFrames
batting_data = pd.read_csv('/kaggle/input/cricket/Batting_data.csv')
bowling_data = pd.read_csv('/kaggle/input/cricket/Bowling_data.csv')
fielding_data = pd.read_csv('/kaggle/input/cricket/Fielding_data.csv')
fantasy_data = pd.read_csv('/kaggle/input/cricket/Final_Fantasy_data.csv')
match_details = pd.read_csv('/kaggle/input/cricket/Match_details.csv')
```

```python
# Merging Batting and Bowling data on match_id, season, fullName
batting_bowling_data = pd.merge(batting_data, bowling_data, 
                                on=['match_id', 'season', 'fullName', 'match_name', 'home_team', 'away_team', 'venue'], 
                                how='outer')

# Now, merge with Fielding data
combined_data = pd.merge(batting_bowling_data, fielding_data, 
                         on=['match_id', 'season', 'fullName', 'match_name', 'home_team', 'away_team', 'venue'], 
                         how='outer')
```

```python
# Merge with Final Fantasy Data
full_data = pd.merge(combined_data, fantasy_data, 
                     on=['match_id', 'season', 'fullName', 'match_name', 'home_team', 'away_team', 'venue'], 
                     how='outer')
```

```python
# Merge with Match Details
final_data = pd.merge(full_data, match_details, 
                      on=['match_id', 'season', 'match_name', 'venue', 'home_team', 'away_team'], 
                      how='outer')
```

```python
# Example: Fill missing fantasy points with 0
final_data.fillna(0, inplace=True)
```

```python
import pandas as pd

# Assuming you have a DataFrame called df with 'batting_team_x' and 'batting_team_y'
final_data['batting_team'] = final_data.apply(
    lambda row: row['batting_team_x'] if pd.notnull(row['batting_team_x']) and row['batting_team_x'] != 0
    else row['batting_team_y'], axis=1)

```

```python



# Replace zeros in the 'team' column with corresponding 'bowling_team' values
final_data['team'] = final_data.apply(lambda row: row['bowling_team'] if row['team'] == 0 else row['team'], axis=1)

# Save the modified DataFrame back to a CSV file
final_data.to_csv('updated_file.csv', index=False)
```

```python

# Create the new 'team' column with the logic:
final_data['team'] = final_data.apply(lambda row: row['batting_team_x'] if row['batting_position'] != 0 else row['bowling_team_y'], axis=1)

# Drop the unnecessary columns
final_data.drop(columns=['batting_team_x', 'batting_team_y', 'bowling_team_x', 'bowling_team_y'], inplace=True)

# Display the updated DataFrame
print(final_data.head())
```
